```python
import networkx as nx
 
g = nx.Graph()
 
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(3, 4)
g.add_edge(1, 4)
g.add_edge(1, 5)
 
nx.draw(g,with_labels=True, font_weight='bold')
```

```python
nx.write_edgelist(g,'Grafos.csv', data=False)
```
